```python
def distance_(cust_1, cust_2):
    return (((cust_1["XCOORD"] - cust_2["XCOORD"])**2 + (cust_1["YCOORD"] - cust_2["YCOORD"])**2)**(1/2))
```

```python
class Truck:
    def __init__(self):
        self.route = []
        self.go_from = 0
        self.go_to = 0
        self.load = capacity
        self.timer = 0
    def __eq__(self, other): 
         return self.go_from == other.go_from and self.go_to == other.go_to
```

```python
def find_start_desicion(customers, ntruks):
    lenght = 0
    trucks = []
    for i in range(ntruks):
        trucks.append(Truck())
    not_visited = customers.shape[0] - 1
    for truck in trucks:
        truck.go_from = customers.iloc[0]
        truck.route.append(customers.iloc[0])
    while(not_visited > 0):
        cant_find_decision = 0
        for truck in trucks:
            posible_distances = []
            for i in range(1, customers.shape[0]):
                cust_finish_time = customers.iloc[i]["DUE DATE"]
                if ((customers.iloc[i]["VISITED"] == False) & (cust_finish_time > truck.timer) & (truck.load >= customers.iloc[i]["DEMAND"])):
                    dist = distance_(truck.go_from, customers.iloc[i])
                    posible_distances.append((dist, i))
            if (len(posible_distances) > 0):
                closest_cust = min(posible_distances, key=lambda x:x[0])
                if (truck.timer < customers.iloc[closest_cust[1]]["DUE DATE"]):
                    truck.go_to = customers.iloc[closest_cust[1]]
                    lenght += closest_cust[0] + truck.go_to["SERVICE TIME"]
                    truck.timer += closest_cust[0] + (truck.go_to["READY TIME"] - truck.timer if (truck.go_to["READY TIME"] > truck.timer) else 0) + truck.go_to["SERVICE TIME"]
                    truck.load -= truck.go_to["DEMAND"]
                    print(not_visited)
                    customers.iloc[closest_cust[1]]["VISITED"] = True
                    truck.route.append(truck.go_to)
                    truck.go_from = truck.go_to
                    not_visited -= 1
                    if (not_visited == 0):
                        break
            else:
                cant_find_decision += 1
                print("ERROR: ",cant_find_decision)
        if (cant_find_decision == ntrucks):
            return 0, 0
    
    routes = []
    for truck in trucks:
        truck.route.append(customers.iloc[0])
        routes.append(truck.route)
    return routes, lenght
```

```python
def find_min_lories(customers):
    sum_demands = 0
    for index, customer in customers.iterrows():
        sum_demands += customer["DEMAND"]
    num_lories = int(sum_demands / capacity) + 1
    return num_lories
```

```python
to_change_cust = customers.copy()
ntrucks = find_min_lories(customers)
ntrucks = int(customers.shape[0] / 2)

print(ntrucks) ## 43 для 600 кастомеров

routes, lenght = find_start_desicion(to_change_cust, ntrucks)
while(lenght == 0):
    clear_output()
    print(ntrucks, " failed")
    to_change_cust = customers.copy()
    routes, lenght = find_start_desicion(to_change_cust, ntrucks)
    ntrucks += 1
print(lenght)
```

```python
def total_distance(routes___):
    tot_distance = 0
    for i_1 in range(len(routes___)):
        for j_1 in range(len(routes___[i_1]) - 1):
            tot_distance += distance_(routes___[i_1][j_1], routes___[i_1][j_1 + 1])
    return tot_distance
```

```python
def count_route_dist(route):
    dist = 0
    for i_3 in range(len(route) - 1):
        dist += distance_(route[i_3], route[i_3 - 1])
    return dist
```

```python
def aprove_solution(routes):
    ncusts_must = customers.shape[0]
    ncusts_now = 0
    for route in routes:
        timer = 0
        load = capacity
        for ind in range(len(route) - 1):
            if ((route[ind]["DUE DATE"] < timer) | (load < 0)):
                return False
            else:
                timer += distance_(route[ind], route[ind + 1]) + route[ind + 1]["SERVICE TIME"] + (timer - route[ind + 1]["READY TIME"] if (route[ind + 1]["READY TIME"] > timer) else 0)
                load -= route[ind]["DEMAND"]
                ncusts_now += 1
    ncusts_now -= len(routes)
    print(ncusts_must, ncusts_now)
    return True
```

```python
def aprove_route_solution(route_):
    timer = 0
    for ind in range(len(route_) - 1):
        if ((route_[ind]["DUE DATE"] < timer)):
            return False
        else:
            timer += distance_(route_[ind], route_[ind + 1]) + route_[ind + 1]["SERVICE TIME"] + (timer - route_[ind + 1]["READY TIME"] if (route_[ind + 1]["READY TIME"] > timer) else 0)
    return True
```

```python
def to_best_place(pt, route, rts_, rts_dist, ind):
    load = capacity
    for cust in route:
        load -= cust["DEMAND"]
    load -= pt["DEMAND"]
    if (load < 0):
        print("Not anought of load")
        return route, False
    t_route = route.copy()
    
    for cust_ind in range(1, len(route)):
        t_route.insert(cust_ind, pt)
        if (aprove_route_solution(t_route) == True):
            rts_tmp = copy.deepcopy(rts_)
            rts_tmp[ind] = t_route
            tmp_dist = total_distance(rts_tmp)
#             print(total_distance(rts_), tmp_dist, rts_dist)
            if (tmp_dist < rts_dist):
                return t_route, True
        t_route = route.copy()
    return route, False
```

```python
import random
def local_search(_routes_):
    range_i = [random.randrange(1, len(_routes_), 1) for i in range(int(len(_routes_)/3))]
    print(range_i)
    range_k = [random.randrange(1, len(_routes_), 1) for i in range(int(len(_routes_)/3))]
    print(range_k)
    for i in range_i:
        for k in range_k:
            if (i == k):
                continue
            for j in range(1, len(_routes_[i]) - 1):
                from_delete_tmp = _routes_[i].copy()
                from_delete = _routes_[i].copy()
                pt_ = from_delete[j].copy()
                abs_dist = total_distance(_routes_)
                del(from_delete[j])

                to_insert = _routes_[k].copy()
                tmp_rts_ = copy.deepcopy(_routes_)
                tmp_rts_[i] = from_delete
                to_insert, is_changed = to_best_place(pt_, to_insert, tmp_rts_, abs_dist, k)
                if (is_changed == True):
                    tmp_rts_[k] = to_insert
                    _routes_ = copy.deepcopy(tmp_rts_)
                    break
                else:
                    _routes_[i] = from_delete_tmp
    return _routes_
```

```python
def two_opt_swap(route):
    route_dist = count_route_dist(route)
    for i in range(1,len(route) - 1):
        route_tmp = route[0: i]
        route_tmp.extend([route[z] for z in range(len(route) - i - 1, i - 1, -1)])
        route_tmp.extend(route[len(route) - i: len(route)])
        if (count_route_dist(route_tmp) < count_route_dist(route)):
            return route_tmp
    return route
```

```python
def perturb(routes):
    for _i_ in range(len(routes)):
        res = two_opt_swap(routes[_i_])
        if (count_route_dist(res) < count_route_dist(routes[_i_])):
#             for asdasdaksd in routes[_i_]:
#                 print(asdasdaksd["CUST NO"])
#             for asdasdaksd in res:
#                 print(asdasdaksd["CUST NO"])
            routes[_i_] = res
    return routes
```

```python
def get_dist_matrix(customers):
    dist_mat = []
    for i in range(customers.shape[0]):
        dist_cust = []
        for j in range(customers.shape[0]):
            if i == j:
                dist_cust.append(0)
            else:
                dist_cust.append(distance_(customers.iloc[i], customers.iloc[j]))
        dist_mat.append(dist_cust)
    return dist_mat
```

```python
def get_indicators(customers, routes):
    inds = [ [ 0 for j in range(customers.shape[0])] for i in range(customers.shape[0])]
    for i in range(len(routes)):
        for j in range(len(routes[i]) - 1):
            inds[routes[i][j]["CUST NO"]][routes[i][j + 1]["CUST NO"]] = 1
    return inds
```

```python
def get_eval(routes, alpha, indcators, penalties, n_customers):
    val = 0
    for i in range(n_customers):
        for j in range(n_customers):
            val+= indcators[i][j] * penalties[i][j] 
    res = total_distance(routes) + alpha * val
    return res
```

```python
def get_utils(indcators, dist_matrix, penalties, n_customers):
    utils_ = []
    for i in range(n_customers):
        util = []
        for j in range(n_customers):
            util.append(indcators[i][j] * dist_matrix[i][j] / (1 + penalties[i][j]))
        utils_.append(util)
    return utils_
```

```python
def update_penalties(penalties, utils):
    max_s = []
    for i in range(len(penalties)):
        max_pen_ind = penalties[i].index(max(penalties[i]))
        max_s.append((i, max_pen_ind))
    for i in range(len(max_s)):
        for j in max_s[i]:
            penalties[i][j] = penalties[i][j] + 1
    return penalties
```
