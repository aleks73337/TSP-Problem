```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output

txt_path = "instances/C203.txt"
file = open(txt_path, 'r')
import re
number = 0
capacity = 0
counter = 0
customers = pd.DataFrame(columns = ['CUST NO', 'XCOORD', 'YCOORD', 'DEMAND', 'READY TIME', 'DUE DATE', 'SERVICE TIME', 'VISITED'])
for line in file:
    if (counter == 4):
        values = re.findall("\s*\d*\s*", line)
        number = int(values[0].replace(" ", ""))
        capacity = int(values[1].replace(" ", ""))
    if (counter > 8):
        values = re.findall("\s*\d*\s*", line)
        customers = customers.append({'CUST NO' : int(values[0].replace(" ", "")), 
                          'XCOORD' : int(values[1].replace(" ", "")), 
                          'YCOORD' : int(values[2].replace(" ", "")), 
                          'DEMAND' : int(values[3].replace(" ", "")), 
                          'READY TIME' : int(values[4].replace(" ", "")), 
                          'DUE DATE' : int(values[5].replace(" ", "")), 
                          'SERVICE TIME' : int(values[6].replace(" ", "")),
                          'VISITED': False}, ignore_index=True)
    counter = counter + 1
file.close()
```

```python
def distance_(cust_1, cust_2, time):
    return (((cust_1["XCOORD"] - cust_2["XCOORD"])**2 + (cust_1["YCOORD"] - cust_2["YCOORD"])**2)**(1/2)) + (cust_2["READY TIME"] - time if cust_2["READY TIME"] > time else 0)
```

```python
class Truck:
    def __init__(self):
        self.route = []
        self.go_from = 0
        self.go_to = 0
        self.load = capacity
        self.timer = 0
    def __eq__(self, other): 
         return self.go_from == other.go_from and self.go_to == other.go_to
```

```python
def find_start_desicion(customers, ntruks):
    lenght = 0
    trucks = []
    for i in range(ntruks):
        trucks.append(Truck())
    not_visited = customers.shape[0] - 1
    for truck in trucks:
        truck.go_from = customers.iloc[0]
    while(not_visited > 0):
        cant_find_decision = 0
        for truck in trucks:
            posible_distances = []
            for i in range(1, customers.shape[0]):
                cust_finish_time = customers.iloc[i]["DUE DATE"]
                if ((customers.iloc[i]["VISITED"] == False) & (cust_finish_time > truck.timer) & (truck.load >= customers.iloc[i]["DEMAND"])):
                    dist = distance_(truck.go_from, customers.iloc[i], truck.timer)
                    posible_distances.append((dist, i))
            if (len(posible_distances) > 0):
                closest_cust = min(posible_distances, key=lambda x:x[0])
                if (truck.timer < customers.iloc[closest_cust[1]]["DUE DATE"]):
                    truck.go_to = customers.iloc[closest_cust[1]]
                    lenght += closest_cust[0] + truck.go_to["SERVICE TIME"]
                    truck.timer += closest_cust[0] + truck.go_to["SERVICE TIME"]
                    truck.load -= truck.go_to["DEMAND"]
                    print(not_visited)
                    customers.iloc[closest_cust[1]]["VISITED"] = True
                    truck.route.append((truck.go_from, truck.go_to))
                    truck.go_from = truck.go_to
                    not_visited -= 1
                    if (not_visited == 0):
                        break
            else:
                cant_find_decision += 1
                print("ERROR: ",cant_find_decision)
        if (cant_find_decision == ntrucks):
            return 0, 0
    
    routes = []
    for truck in trucks:
        truck.route.append((truck.go_to, customers.iloc[0]))
        routes.append(truck.route)
    return routes, lenght
```

```python
def find_min_lories(customers):
    sum_demands = 0
    for index, customer in customers.iterrows():
        sum_demands += customer["DEMAND"]
    num_lories = int(sum_demands / capacity) + 1
    return num_lories
```

```python
to_change_cust = customers.copy()
ntrucks = find_min_lories(customers)

print(ntrucks) ## 43 для 600 кастомеров

routes, lenght = find_start_desicion(to_change_cust, ntrucks)
while(lenght == 0):
    clear_output()
    print(ntrucks, " failed")
    to_change_cust = customers.copy()
    routes, lenght = find_start_desicion(to_change_cust, ntrucks)
    ntrucks += 1
print(lenght)
```

```python
def find_unique(route):
    unique = [route[i][0] for i in range(len(route))]
    return unique
```

```python
def swap(step, cust, route):
    cust_to = 0
    for i in range(len(route)): ## выбираем шаг в котором "куда" это наш кастомер
        if route[i][1]["CUST NO"] == cust["CUST NO"]:
            cust_to = route[i]
    step_1 = (step[0], cust_to[1])
    step_2 = (cust_to[0], step[1])
    return step_1, step_2, cust_to
```

```python
def find_distance(custs):
    distance = 0
    for i in range(len(custs) - 1):
        distance += distance_(custs[i], custs[i+1], distance) + custs[i+1]["DEMAND"] + custs[i+1]["SERVICE TIME"]
    return distance
```

```python
def _2optSwap(route, i, k):
    new_route = route[0: i]
    for j in range(k, i - 1, -1):
        new_route.append(route[j])
    for l in range(k+1, len(route)):
        new_route.append(route[l])
    return(new_route)
```

```python
def local_search(route):
    depot = route[0][0]
    custs = find_unique(route)
    custs = custs[1:len(custs)]
    best_distance = find_distance(custs)
    prev_dist = best_distance
    cant_improve = (len(custs)**2)/3
    while(cant_improve > 0):
        for i in range(0, len(custs)-1):
            for k in range(i+1, len(custs)):
                new_route = _2optSwap(custs, i, k)
                new_distance = find_distance(new_route)
                if (new_distance < best_distance):
                    custs = new_route
                    best_distance = new_distance
                else:
                    cant_improve -= 1
    custs.insert(0, depot)
    custs.append(depot)
    print(prev_dist)
    print(best_distance)
    return custs
```

```python
def local_seatch_improve(routes):
    for route in routes:
        plot_way(route).show()
        route = local_search(route)
        way = [(route[i], route[i+1]) for i in range(len(route) - 1)]
        plot_way(way).show()
```
